```python
###comparison of LGAR and LGARTO results

#LGARTO can be run in LGAR mode or LGARTO mode, simply by changing whether or not the 
#boolean TO_enabled is set to true or false in the congig file. 
#In the development of LGARTO, very small changes were necessary to some parts of the LGAR
#code itself. This notebook demonstrates that these differences are very small, such that 
#LGAR as used by setting TO_enabled = false in the LGARTO code is essentially the same as 
#LGAR before the TO developments. In some cases the results are identical. 
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
```

```python
c_path_vars = 'outputs/synthetic1/data_variables.csv'

data_c_vars = pd.read_csv(c_path_vars)
```

```python
c_path_vars_LGARTO = 'outputs/synthetic1_LGARTO/data_variables.csv'

data_c_vars_LGARTO = pd.read_csv(c_path_vars_LGARTO)
```

```python
hours_vec = []

for i in range(0,len(data_c_vars_LGARTO)):
    delta = datetime.datetime.strptime(data_c_vars_LGARTO['Time'][i], '%Y-%m-%d %H:%M:%S') - datetime.datetime.strptime(data_c_vars_LGARTO['Time'][0],'%Y-%m-%d %H:%M:%S') 
    hour_delta = delta.total_seconds()/3600
    hours_vec.append(hour_delta)
    
```

```python
plt.plot(hours_vec, data_c_vars_LGARTO['surface_runoff'], label = 'LGARTO runoff')
plt.plot(hours_vec, data_c_vars['surface_runoff'], label = 'LGAR runoff')
plt.plot(hours_vec, data_c_vars['surface_runoff'] - data_c_vars_LGARTO['surface_runoff'], label = 'difference')
plt.ylabel("flux (m/h)")
plt.xlabel("time (h)")
plt.text(s="synth 1", x=2, y=0.0004)
plt.legend()
```

```python
plt.plot(hours_vec, data_c_vars_LGARTO['infiltration'], label = 'LGARTO infiltration')
plt.plot(hours_vec, data_c_vars['infiltration'], label = 'LGAR infiltration')
plt.plot(hours_vec, data_c_vars['infiltration'] - data_c_vars_LGARTO['infiltration'], label = 'difference')
plt.ylabel("flux (m/h)")
plt.xlabel("time (h)")
plt.text(s="synth 1", x=2, y=0.0004)
plt.legend()
```

```python
cumulative_infiltration_vec = []
cumulative_infiltration = 0.0
cumulative_runoff_vec = []
cumulative_runoff = 0.0

for i in range(0,len(data_c_vars_LGARTO)):
    cumulative_infiltration_vec.append(np.sum(data_c_vars_LGARTO['infiltration'][0:i]))
    cumulative_runoff_vec.append(np.sum(data_c_vars_LGARTO['surface_runoff'][0:i]))

data_c_vars_LGARTO['cumulative_infiltration'] = cumulative_infiltration_vec
data_c_vars_LGARTO['cumulative_runoff'] = cumulative_runoff_vec


cumulative_infiltration_vec = []
cumulative_infiltration = 0.0
cumulative_runoff_vec = []
cumulative_runoff = 0.0

for i in range(0,len(data_c_vars)):
    cumulative_infiltration_vec.append(np.sum(data_c_vars['infiltration'][0:i]))
    cumulative_runoff_vec.append(np.sum(data_c_vars['surface_runoff'][0:i]))

data_c_vars['cumulative_infiltration'] = cumulative_infiltration_vec
data_c_vars['cumulative_runoff'] = cumulative_runoff_vec
```

```python
(data_c_vars['cumulative_infiltration'] - data_c_vars_LGARTO['cumulative_infiltration']).tail()
```

```python
(data_c_vars['cumulative_runoff'] - data_c_vars_LGARTO['cumulative_runoff']).tail()
```

```python
#we can see that for the first synthetic simulation, 
#there is no difference between LGAR and LGARTO.
```
